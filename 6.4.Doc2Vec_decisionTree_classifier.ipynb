### 4. TFIDF decision Tree classifier

```python
import pandas as pd

lyrics = pd.read_csv('../dataset/Lyrics_en_clean.csv', index_col=['Band','Song'])
genres = pd.read_csv('../dataset/artists_final_ohe.csv', index_col=['Band'])

#df = pd.concat([lyrics, genres], axis=1, join_axes=[lyrics.index], levels=lyrics.index.levels)
df = pd.merge(lyrics, genres, on=['Band'])
del lyrics
del genres
df
```

```python
y_train = df.drop(['Lyrics'], axis=1)
del df
```

```python
from gensim.models.doc2vec import Doc2Vec

doc2vec = Doc2Vec.load("../dataset/doc2vec")
doc_vectors = doc2vec.wv
print("Number of doc vectors: {}".format(len(doc_vectors.vocab)))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(doc2vec.docvecs,y_train, test_size=0.4)

print('Train shape: ', len(X_train), ",", y_train.shape)
print('Test shape: ', len(X_test), ",", y_test.shape)
```

```python
%%time
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

```python
from sklearn.externals import joblib

joblib.dump(clf, '../dataset/decisionTree_doc2vec') 
```

```python
%%time
from sklearn.externals import joblib
from sklearn.metrics import accuracy_score

clf = joblib.load('../dataset/decisionTree_doc2vec')
y_pred = clf.predict(X_test)
print(accuracy_score(y_test, y_pred))           #This returns 0 score
```
