## 2. Sentiment Analysis 
Use Vader sentiment analyzer, it was trained using online reviews.

This library returns a value between -1 and 1, the bigger the value the more positive the song. 0 means neutral.

```python
import pandas as pd
df = pd.read_csv('../dataset/Lyrics_en_clean.csv')
df['sentiment'] = '0'
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyser = SentimentIntensityAnalyzer()

for row in df.itertuples(index=True, name='Pandas'):
    lyric = getattr(row, "Lyrics")
    index = getattr(row, "Index")
    try:
        scores = analyser.polarity_scores(lyric)
        df.at[index, 'sentiment'] = scores['compound']
    except:
        pass
    
df = df.drop(columns="Lyrics")
df.to_csv('../dataset/song_sentiment.csv', index=False)
```
