### 2. Delete non english lyrics


```python
from langdetect import detect
from langdetect import DetectorFactory 
DetectorFactory.seed = 0

lyrics_dropped = 0
for row in df.itertuples(index=True, name='Pandas'):
    lyric = getattr(row, "Lyrics")
    index = getattr(row, "Index")
    try:
        if detect(lyric) != 'en':
            df.drop(index, inplace=True)
            lyrics_dropped += 1
    except:
        pass

print("Lyrics dropped ", lyrics_dropped)
df.to_csv('../dataset/Lyrics_en.csv', index=False)
df
```

```python
import ast
df2 = pd.read_csv('../dataset/artists.csv')

genres = set([])
for row in df2.itertuples(index=True, name='Pandas'):
    value = ast.literal_eval(getattr(row, "Genres"))
    for i in range(len(value)):
        value2 = value[i].split(" ")
        for j in range(len(value2)):
            genres.add(value2[j])
print("Unique genres: ", len(genres))
#print(genres)
df2
```
